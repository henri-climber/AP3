```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# 1) Messwerte (von dir)
# -----------------------------
# Brechungsindizes n und Unsicherheiten Δn
n = np.array([1.622, 1.625, 1.643])
dn = np.array([0.002, 0.002, 0.002])

# Wellenlängen λ in nm:
# HIER deine Literaturwerte eintragen (Hg-Lampe / Spektrallinien)!
# Beispielwerte (bitte ggf. ersetzen):
lam = np.array([579.0, 546.1, 435.8])   # orange, grün, blau
# -----------------------------

# -----------------------------
# 2) Linearen Fit machen
#    Fit-Form: n(λ) = a*λ + b
# -----------------------------
a, b = np.polyfit(lam, n, 1)

# Fit-Linie zum Zeichnen
lam_fit = np.linspace(lam.min() - 20, lam.max() + 20, 300)
n_fit = a * lam_fit + b

# -----------------------------
# 3) Plot: λ gegen n + Fehlerbalken + Fit
# -----------------------------
plt.errorbar(lam, n, yerr=dn, fmt='o', markerfacecolor = 'none', capsize=4, label='Messwerte')
plt.plot(lam_fit, n_fit, label=f'Linearer Fit: n = {a:.3e}·λ + {b:.4f}')

plt.xlabel(r'Wellenlänge $\lambda$ [nm]')
plt.ylabel(r'Brechungsindex $n$')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

print("Fitparameter:")
print(f"a = {a:.6e} 1/nm")
print(f"b = {b:.6f}")

# -----------------------------
# 4) Aus dem Fit λ für gegebene n berechnen
#    Umstellen: n = a*λ + b  ->  λ = (n - b)/a
# -----------------------------
lam_from_fit = (n - b) / a

print("\nAus dem Fit berechnete Wellenlängen (für deine n-Werte):")
labels = ["orange", "grün", "blau"]
for lab, lam_i, n_i in zip(labels, lam_from_fit, n):
    print(f"{lab:6s}: n = {n_i:.3f} -> λ_fit = {lam_i:.1f} nm")




```
