```python
import math

lbd = math.log(2) / (30.2 * 365.25 * 24 * 3600)
```

```python
lbd
```

```python
A = 37*1000 * math.exp(- lbd * 28 *365.25 * 24 * 3600)
```

```python
A
```

```python
raumwinkel = 0.72 #sr
```

```python

# Berechnung der Energieübertragung pro Jahr
# Cs-137 gibt γ-Quanten mit E_γ = 0.662 MeV ab

E_gamma = 0.662  # MeV
E_gamma_joule = E_gamma * 1.602e-13  # Umwandlung in Joule
abstand = 1  # m
full_sphere = 4 * math.pi  # Raumwinkel einer vollständigen Sphäre

# Bruchteil der Strahlung, die auf den Körper trifft
fraction_body = raumwinkel / full_sphere

# Anzahl der γ-Quanten pro Jahr
sekunden_pro_jahr = 365.25 * 24 * 3600
anzahl_quanten_pro_jahr = A * sekunden_pro_jahr

# Energie auf den Körper pro Jahr
energie_pro_jahr_joule = anzahl_quanten_pro_jahr * E_gamma_joule * fraction_body
energie_pro_jahr_gy = energie_pro_jahr_joule / 70  # 1 Gy = 1 J/kg, Körpergewicht 70 kg
```

```python
print(f"Aktivität 2025: {A:.2f} Bq")
print(f"Raumwinkel des Körpers: {raumwinkel} sr")
print(f"Anteil der Strahlung auf den Körper: {fraction_body:.4e}")
print(f"Anzahl γ-Quanten pro Jahr: {anzahl_quanten_pro_jahr:.2e}")
print(f"Energieübertragung pro Jahr: {energie_pro_jahr_joule:.2e} J")
print(f"Energiedosis (Gy): {energie_pro_jahr_gy:.4f} Gy = {energie_pro_jahr_gy*100:.4f} rad")
```

```python

# Effektive Dosis
# Qualitätsfaktor Q für γ-Strahlung: Q = 1
# Effektive Dosis H = Q * D (in Sv bzw. rem)

Q_gamma = 1  # Qualitätsfaktor für Gammastrahlung
effektive_dosis_sv = energie_pro_jahr_gy * Q_gamma
effektive_dosis_msv = effektive_dosis_sv * 1000
effektive_dosis_rem = effektive_dosis_msv / 10
```

```python
print(f"\nEffektive Dosis:")
print(f"  {effektive_dosis_sv:.4e} Sv")
print(f"  {effektive_dosis_msv:.4f} mSv")
print(f"  {effektive_dosis_rem:.4f} rem")
print(f"\nZum Vergleich: Natürliche jährliche Strahlenbelastung ~ 2-3 mSv")
```

```python
# Plausibilitätscheck mit Dosisleistungskonstante (Gamma-Konstante)
# Für Cs-137: Gamma ~= 0.08 uSv * m^2 / (h * MBq)
# Dies berücksichtigt, dass nicht alles absorbiert wird (realistischerer Wert)

gamma_konstante = 0.08 # uSv * m^2 / (h * MBq)
aktivitaet_mbq = A / 1e6 # Umrechnung in MBq

dosisleistung_pro_h = gamma_konstante * aktivitaet_mbq / (abstand**2) # uSv/h
dosis_jahr_realistisch = dosisleistung_pro_h * 24 * 365.25 / 1000 # mSv

print("\n--- PLAUSIBILITÄTSCHECK ---")
print(f"Ihre Berechnung (Worst-Case, 100% Absorption): {effektive_dosis_msv:.4f} mSv")
print(f"Über Gamma-Konstante (Realistische Wechselwirkung): {dosis_jahr_realistisch:.4f} mSv")
print("\nBewertung: Der Wert ist realistisch und sehr klein.")
print("Die Strahlenbelastung liegt im Bereich von ca. 1-2% der natürlichen Hintergrundstrahlung.")
print("37 kBq ist eine typische 'Freigrenze' für schulische Unterrichtsmittel.")
```
