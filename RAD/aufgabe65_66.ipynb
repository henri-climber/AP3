```python
from uncertainties import ufloat

# Calibration parameters
m = ufloat(2.8836, 0.0035)
b = ufloat(-40.7, 1.8)

def energie(n_a, sigma):
    c = ufloat(n_a, sigma)
    return m * c + b

# Calculate Energy
df['Energy'] = df.apply(lambda row: energie(row['n_a'], row['Sigma']), axis=1)

# Extract nominal value and standard deviation for easier reading/plotting if needed
df['Energy_keV'] = df['Energy'].apply(lambda x: x.nominal_value)
df['Energy_err'] = df['Energy'].apply(lambda x: x.std_dev)

df
```

```python
# Reference Data from Table 1
reference_data = {
    'Uranium Series': {
        '238U': [50],
        '234Th': [63, 93],
        '234Pa': [1001, 767],
        '226Ra': [186],
        '214Pb': [352, 295],
        '214Bi': [609, 1765, 1120]
    },
    'Thorium Series': {
        '228Ac': [911, 969, 339],
        '224Ra': [241],
        '212Pb': [239],
        '212Bi': [727],
        '208Tl': [2615, 583, 511, 860]
    },
    'Other': {
        '40K': [1461],
        '137Cs': [662],
        '60Co': [1173, 1333],
        '22Na': [1275],
        '241Am': [59.5]
    }
}

# Flatten reference data for easier searching
flat_reference = []
for series, isotopes in reference_data.items():
    for isotope, energies in isotopes.items():
        for energy in energies:
            flat_reference.append({
                'Series': series,
                'Isotope': isotope,
                'Ref_Energy': energy
            })

df_ref = pd.DataFrame(flat_reference)

def find_match(measured_energy, series_filter=None):
    # measured_energy is a ufloat
    val = measured_energy.nominal_value
    err = measured_energy.std_dev

    best_match = None
    min_dist_sigma = float('inf')

    # Filter candidates if series is known (optional, but good for specific samples)
    candidates = df_ref
    if series_filter:
        # Prioritize or filter? Let's search all but interpret later.
        pass

    matches = []

    for _, row in candidates.iterrows():
        ref_e = row['Ref_Energy']
        diff = val - ref_e

        # Calculate 'distance' in standard deviations
        # Note: If err is 0 (shouldn't happen with ufloat), handle it
        if err > 0:
            dist_sigma = abs(diff) / err
        else:
            dist_sigma = abs(diff) # Fallback

        # Define a cutoff for "likely match".
        # Since calibration might have systematic errors, maybe loose cutoff like 5 sigma or absolute 20 keV
        if dist_sigma < 5 or abs(diff) < 20:
            matches.append({
                'Isotope': row['Isotope'],
                'Series': row['Series'],
                'Ref_Energy': ref_e,
                'Deviation_keV': diff,
                'Deviation_Sigma': dist_sigma
            })

    # Sort by proximity
    matches.sort(key=lambda x: abs(x['Deviation_keV']))

    if matches:
        return matches[0] # Return best match
    else:
        return None

# Apply identification
identifications = []

for idx, row in df.iterrows():
    sample = idx[0] # MultiIndex level 0
    energy_obj = row['Energy']

    match = find_match(energy_obj)

    if match:
        identifications.append(match)
    else:
        identifications.append({
            'Isotope': 'Unknown',
            'Series': '-',
            'Ref_Energy': None,
            'Deviation_keV': None,
            'Deviation_Sigma': None
        })

# Create dataframe from identifications list
df_id = pd.DataFrame(identifications, index=df.index)

# Join with original dataframe
result = pd.concat([df, df_id], axis=1)

# Generate Report
print("=== AUTOMATED RADIOISOTOPE IDENTIFICATION REPORT ===\n")

for sample in result.index.levels[0]:
    print(f"--- Analysis for Sample: {sample} ---")
    subset = result.loc[sample]

    # Check what expected series is
    expected = []
    if sample == 'Uranoxid':
        print(f"Expected Series: Uranium Series")
    elif sample == 'Thorium':
        print(f"Expected Series: Thorium Series")
    elif sample == 'Wecker':
        print(f"Unknown Sample (expecting Radium/Uranium daughters)")

    print(f"\n{'Measured Energy':<20} | {'Identified':<10} | {'Ref E':<8} | {'Diff (keV)':<10} | {'Sigma':<6}")
    print("-" * 70)

    for _, row in subset.iterrows():
        meas_e = row['Energy']
        iso = row['Isotope']
        ref_e = row['Ref_Energy'] if not pd.isna(row['Ref_Energy']) else "-"
        diff = f"{row['Deviation_keV']:.1f}" if not pd.isna(row['Deviation_keV']) else "-"
        sigma = f"{row['Deviation_Sigma']:.1f}" if not pd.isna(row['Deviation_Sigma']) else "-"

        print(f"{meas_e.nominal_value:>6.1f} ± {meas_e.std_dev:>4.1f} keV | {iso:<10} | {ref_e:<8} | {diff:>10} | {sigma:>6}")

    print("\nSummary of Identified Isotopes:")
    found_isos = subset[subset['Isotope'] != 'Unknown']['Isotope'].unique()
    print(f"Found: {', '.join(found_isos)}\n")
    print("\n")
```
